## EP 1 - Cálculos Complexos com Séries de Taylor

Formula escolhida: y = x³ − 10x² + 6

```python
import numpy as np
import matplotlib.pyplot as plt
import math

#Valores de X
x = np.linspace(-20,20,40)
```

### Derivada 
Sendo a formula y = x³ - 10x² + 6, um polinomio, ao se aplicar a série de Taylor, acabamos retornonando para a mesma equação.  

$f(x) = Σ^∞_{n = 0} {(f(a))^n \over n!}(x-a)^n$

$f(x)  =f(a)  + {f'(a) \over 1!}(x-a) + {f''(a) \over 2!}(x-a)^2 + {f'''(a) \over 3!}(x-a)^3 + ... $

$6 + {{d \over dx}(x^3-10x^2+6)(0) \over 1!}x + {{d^2 \over dx^2}(x^3-10x^2+6)(0) \over 2!}x^2 + {{d^3 \over dx^3}(x^3-10x^2+6)(0) \over 3!}x^3 + ... $

$6 + {0 \over 1!}x + {-20 \over 2!}x^2 + {6 \over 3!}x^3 +... $

$6-10x^2 + x^3$

```python
#Formula original
def polinomio_orig(x):
  return 6-(10*(x**2))+(x**3)

#Formula simplificada
def polinomio_taylor(x):
  return 6-(10*(x**2))+(x**3)

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(5, 3))
fig.set_size_inches(20,8)

ax[0].plot(x, polinomio_orig(x), linewidth=2.0, label = "Original", color="orange")
ax[1].plot(x, polinomio_taylor(x), linewidth=2.0, label= "Serie Taylor")

ax[0].legend()
ax[1].legend()
plt.show()
```

### Valores de N

Com base no resultado encontrado na série de Taylor temos que N só precisa chegar até o valor 3, pois quando chegamos a valores maiores, a derivada do polinomio da 0, tornando indiferente o N utilizado.

${f(a)} = x³ − 10x² + 6 => {f''''(a)} = 0$

Mas ao aplicar valores menores a N teremos:

```python
#Variando N
def polinomio_nvariavel(x, n=3):
  if(n == 1): 
    return 6-10*x
  return 6+(x**(n-1))*((x**(n-2))-10)

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(5, 3))
fig.set_size_inches(20,8)

ax[0,0].plot(x, polinomio_orig(x), linewidth=2.0, label = "Original", color="red")
ax[0,1].plot(x, polinomio_nvariavel(x), linewidth=2.0, label= "N=3")
ax[1,0].plot(x, polinomio_nvariavel(x,2), linewidth=2.0, label= "N=2")
ax[1,1].plot(x, polinomio_nvariavel(x,1), linewidth=2.0, label= "N=1")

ax[0,0].legend()
ax[0,0].grid()
ax[0,1].legend()
ax[0,1].grid()
ax[1,0].legend()
ax[1,0].grid()
ax[1,1].legend()
ax[1,1].grid()

plt.show()
```
