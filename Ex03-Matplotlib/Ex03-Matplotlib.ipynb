### Ex03 Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

mes = np.arange(12) + 1 # meses de 1 a 12

# quantidades de produtos vendidos por mês
creme_facial = np.array([ 2500, 2630, 2140, 3400, 3600, 2760, 2980, 3700, 3540, 1990, 2340, 2900 ])
limpeza_facial = np.array([ 1500, 1200, 1340, 1130, 1740, 1555, 1120, 1400, 1780, 1890, 2100, 1760 ])
pasta_dentaria = np.array([ 5200, 5100, 4550, 5870, 4560, 4890, 4780, 5860, 6100, 8300, 7300, 7400 ])
sabonete = np.array([ 9200, 6100, 9550, 8870, 7760, 7490, 8980, 9960, 8100, 10300, 13300, 14400 ])
shampoo = np.array([ 1200, 2100, 3550, 1870, 1560, 1890, 1780, 2860, 2100, 2300, 2400, 1800 ])
hidratante = np.array([ 1500, 1200, 1340, 1130, 1740, 1555, 1120, 1400, 1780, 1890, 2100, 1760 ])
```

#### Gráfico 1 - Total de produtos vendidos por mês  - Linha

```python
fig_grf1, ax_grf1 = plt.subplots()
fig_grf1.set_size_inches(12, 8)

# Cria o array total mes, ja contendo a soma dos elementos dos arrays dos produtos
total_mes = np.array(creme_facial + limpeza_facial + pasta_dentaria + sabonete + shampoo + hidratante)

ax_grf1.plot(mes, total_mes, label="Total vendido")

ax_grf1.set_xlabel("Mês")
ax_grf1.set_ylabel("Quantidade")
ax_grf1.set_title("Gráfico 1 - Total de produtos Vendidos por mês - Linha")
ax_grf1.legend()

ax_grf1.set(xlim=(1, 12), xticks=(mes))

plt.show()
```

#### Gráfico 2 - Gráfico com todos os produtos vendidos por mês - Linhas

```python
fig_grf2, ax_grf2 = plt.subplots()
fig_grf2.set_size_inches(12, 8)

ax_grf2.plot(mes, creme_facial, label="Creme Facial")
ax_grf2.plot(mes, limpeza_facial, label="Limpeza Facial")
ax_grf2.plot(mes, pasta_dentaria, label="Pasta Dentaria")
ax_grf2.plot(mes, sabonete, label="Sabonete")
ax_grf2.plot(mes, shampoo, label="Shanmpoo")
ax_grf2.plot(mes, hidratante, label="Hidratante")

ax_grf2.set_xlabel("Mês")
ax_grf2.set_ylabel("Quantidade")
ax_grf2.set_title("Gráfico 2 -  Gráfico com todos os produtos vendidos por mês - Linha")
ax_grf2.legend()

ax_grf2.set(xlim=(1, 12), xticks=(mes), ylim=(0, 15000))

plt.show()
```

#### Gráfico 3 - Comparativo de Creme Facial com Limpeza Facial por mês - Barras

```python
fig_grf3, ax_grf3 = plt.subplots()
fig_grf3.set_size_inches(12, 8)

largura_barra = 0.30
ax_grf3.bar(mes - largura_barra/2, creme_facial, largura_barra, label="Creme Facial")
ax_grf3.bar(mes + largura_barra/2, limpeza_facial, largura_barra, label="Limpeza Facail")

ax_grf3.set_xlabel("Mês")
ax_grf3.set_ylabel("Quantidade vendida")
ax_grf3.set_title("Gráfico 3 - Comparativo de Creme Facial com Limpeza Facial por mês - Barras")

ax_grf3.set_xticks(mes)

ax_grf3.legend()

plt.show()
```

#### Gráfico 4 - Histograma de quantidade de meses (y) e faixas de quantidades de produtos vendidos (1000-1999, 2000-2999, ...)

```python
fig_grf4, ax_grf4 = plt.subplots()
fig_grf4.set_size_inches(12, 8)

ax_grf4.hist(total_mes, edgecolor="white")

ax_grf4.set_xlim(total_mes[0], total_mes[len(total_mes)-1])

plt.show()
```

#### Gráfico 5 - Pizza. % da quantidade de cada produto vendido no ano

```python
fig_grf5, ax_grf4 = plt.subplots()
fig_grf5.set_size_inches(12, 8)

total_produto = np.array([np.sum(creme_facial), np.sum(limpeza_facial), np.sum(pasta_dentaria), np.sum(sabonete), np.sum(shampoo), np.sum(hidratante)])
text = ["Creme Facial", "Limpeza Facial", "Pasta Dentaria", "Sabonete", "Shampoo", "Hidratante"]
plt.title("Gráfico 5 - Pizza. % da quantidade produtos vendidos no ano em cada produto")

plt.pie(total_produto, labels=text, autopct='%1.1f%%')
plt.show() 
```
