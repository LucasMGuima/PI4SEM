# Ex07 - Regressão Linear
### Estimando o valor de um diamante

```python
#Libs
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.metrics import r2_score

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#Carregar os dados
df_diamonds = pd.read_csv('dados/diamonds.csv')

#Verficar a tabela
df_diamonds
```

#### Checando os dados

```python
#Checando por dados vazios
df_diamonds.isnull().sum()
```

```python
#Checando o tipo das dados
df_diamonds.dtypes
```

```python
df = df_diamonds.groupby(['cut']).count()
df
```

### Treinamento

```python
df_x = df_diamonds[['carat', 'depth', 'table']]
df_y = df_diamonds[['price']]

#Definir dataset de treinamento e teste
treino_x, test_x, treino_y, test_y = train_test_split(df_x, df_y, test_size=0.2)
```

### Calculo da Regreção Linear

```python
reg = linear_model.LinearRegression()

#treino
reg.fit(treino_x, treino_y)
reg.coef_
```

```python
pred_y = reg.predict(test_x)

r2 = r2_score(test_y, pred_y)
r2*100
```
